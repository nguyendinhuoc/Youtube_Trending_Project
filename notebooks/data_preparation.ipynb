```python
import pandas as pd

# ÄÆ°á»ng dáº«n file
file_path = '../data/trending_yt_videos_113_countries.csv'

# List chá»©a dá»¯ liá»‡u
vn_chunks = []
us_chunks = []
chunk_size = 50000  # Giáº£m xuá»‘ng 50k cho an toÃ n hÆ¡n vá»›i mÃ¡y Codespaces

print("ğŸ”„ Äang xá»­ lÃ½ file vá»›i cháº¿ Ä‘á»™ 'Bá» qua lá»—i kÃ½ tá»±'...")

try:
    # --- ÄOáº N CODE ÄÃƒ Sá»¬A ---
    # ThÃªm encoding_errors='replace': Gáº·p kÃ½ tá»± lá»—i sáº½ thay báº±ng dáº¥u há»i (?) chá»© khÃ´ng crash
    # ThÃªm on_bad_lines='skip': DÃ²ng nÃ o há»ng cáº¥u trÃºc thÃ¬ bá» qua luÃ´n
    # ThÃªm lineterminator='\n': Xá»­ lÃ½ lá»—i xuá»‘ng dÃ²ng trong pháº§n mÃ´ táº£ video
    
    reader = pd.read_csv(
        file_path, 
        chunksize=chunk_size, 
        encoding='utf-8',             
        encoding_errors='replace',    # QUAN TRá»ŒNG: Thay tháº¿ kÃ½ tá»± lá»—i
        on_bad_lines='skip',          # QUAN TRá»ŒNG: Bá» qua dÃ²ng lá»—i
        low_memory=False,
        lineterminator='\n'           # GiÃºp Ä‘á»c Ä‘Ãºng cÃ¡c mÃ´ táº£ cÃ³ xuá»‘ng dÃ²ng
    )

    for i, chunk in enumerate(reader):
        # Lá»c láº¥y VN vÃ  US
        vn_chunks.append(chunk[chunk['country'] == 'VN'])
        us_chunks.append(chunk[chunk['country'] == 'US'])
        
        # In tiáº¿n Ä‘á»™ Ä‘á»ƒ biáº¿t mÃ¡y khÃ´ng bá»‹ treo
        if i % 5 == 0:
            print(f"âœ… ÄÃ£ xá»­ lÃ½ chunk thá»© {i}...")

    # Gá»™p láº¡i
    if vn_chunks:
        df_vn = pd.concat(vn_chunks)
        df_vn.to_csv('../data/vn_youtube_trending.csv', index=False)
        print(f"ğŸ‡»ğŸ‡³ ÄÃ£ lÆ°u file VN: {df_vn.shape[0]} dÃ²ng")
    else:
        print("âš ï¸ KhÃ´ng tÃ¬m tháº¥y dá»¯ liá»‡u VN nÃ o.")

    if us_chunks:
        df_us = pd.concat(us_chunks)
        df_us.to_csv('../data/us_youtube_trending.csv', index=False)
        print(f"ğŸ‡ºğŸ‡¸ ÄÃ£ lÆ°u file US: {df_us.shape[0]} dÃ²ng")
    else:
        print("âš ï¸ KhÃ´ng tÃ¬m tháº¥y dá»¯ liá»‡u US nÃ o.")

    print("ğŸ‰ HoÃ n táº¥t! Báº¡n kiá»ƒm tra folder data nhÃ©.")

except Exception as e:
    print(f"\nâŒ Váº«n lá»—i: {e}")
```
